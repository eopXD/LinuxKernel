```python
import matplotlib.pyplot as plt
```

```python
def read_file ( filename, N ) :
    file = open(filename, "r")
    text = file.readlines()
    S = []
    spent = []
    for line in text :
        line = line.rstrip('\n')
        word = line.split(" ")
        n = int(word[1])
        if ( n == N ) :
            S.append(int(word[3]))
            spent.append(float(word[-1]))
    return (S, spent)
```

```python
insert_small_S, insert_small_spent = read_file("insertion.txt", 5000000)
insert_big_S, insert_big_spent = read_file("insertion.txt", 10000000)
```

```python
bubble_small_S, bubble_small_spent = read_file("bubble.txt", 5000000)
bubble_big_S, bubble_big_spent = read_file("bubble.txt", 10000000)
```

```python
plt.plot([i for i in range(13)], insert_small_spent, 's-', color = 'b', label="Insertion Sort")
plt.plot([i for i in range(13)], bubble_small_spent, 'o-', color = 'r', label="Bubble Sort")

plt.title("N = 5000000", x=0.5, y=1.03, fontsize = 20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# 標示x軸(labelpad代表與圖片的距離)
plt.xlabel("S = 2^(x)", fontsize=15, labelpad = 15)
# 標示y軸(labelpad代表與圖片的距離)
plt.ylabel("Time spent (sec)", fontsize=15, labelpad = 20)
plt.legend(loc = "best", fontsize=20)

fig = plt.gcf()
fig.set_size_inches(10, 5)
fig.savefig('small.png', dpi=200)

plt.show()
```
